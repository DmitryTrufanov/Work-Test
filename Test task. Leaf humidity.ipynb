```python
import pandas as pd
```

```python
df = pd.read_excel('station_data.xlsx')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# удалим лишние столбцы "Эталонная эвапотранспирация ET0 [mm]" - мало данных, "Дата / время" - не влияет на целевой признак
df = df.drop(['Дата / время', 'Эталонная эвапотранспирация ET0 [mm]'], axis=1)
```

```python
# проверим корреляцию признаков
# нет, признака имеющего сильную корреляцию с влажностью листа. (около 57% - солнечная радиация и влажность воздуха)
df.corr(method='spearman').style.background_gradient(cmap='coolwarm')
```

## Сделаем предположение, что влажность листа является категорией (дискретная величина)

## Проведем дискрименантный анализ ##

разделим выборку для обучения и тестирования

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.2)
```

```python
y_train, y_test = df_train['Влажность листа [минимум] time'], df_test['Влажность листа [минимум] time']
del df_train['Влажность листа [минимум] time']
del df_test['Влажность листа [минимум] time']
```

```python
# проверим, что наш целевой парамерт отсутствует в датасете
df_train.info()
```

построим модель

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
lda.fit(df_train, y_train)
```

```python
lda.predict(df_test)
```

```python
result = pd.DataFrame([y_test.values, lda.predict(df_test)]).T
result
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, lda.predict(df_test))
```

Модель предсказывает влажность листа с точностью 76,8%

## Сделаем предположение, что влажность листа является непрерывной величиной

## Проведем регрессионный анализ ##

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()  
lr.fit(df_train, y_train)
```

```python
lr.predict(df_test)
```

```python
result_lr = pd.DataFrame([y_test.values, lr.predict(df_test)]).T
result_lr
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
mse = mean_squared_error(y_test, lr.predict(df_test))
mae = mean_absolute_error(y_test, lr.predict(df_test))
print('mse: %.3f, mae: %.3f' % (mse, mae))
```

```python
import math
math.sqrt(mse)
```

```python
Регрессионная модель хуже предсказывает влажность.
```
